```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from webdriver_manager.microsoft import EdgeChromiumDriverManager
from selenium.webdriver.edge.options import Options

options = Options()
options.use_chromium = True
options.add_argument("headless")  # Run in headless mode, without opening a UI
options.add_argument("disable-gpu")
```

```python
url = "https://discourse.mcneel.com/u/quan_li/activity/solved"
# Setup Edge driver
driver = webdriver.Edge(service=Service(EdgeChromiumDriverManager().install()), options=options)

# Open the webpage
driver.get(url)

# Now you can access the page source after dynamic content has loaded
page_source = driver.page_source
page_source
```

```python
from bs4 import BeautifulSoup

# Assuming 'page_source' is the HTML content obtained from Selenium
soup = BeautifulSoup(page_source, 'html.parser')

soup
```
